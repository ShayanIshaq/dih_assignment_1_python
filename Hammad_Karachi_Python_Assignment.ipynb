```python
columns=['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

#print(dataSet['age'])
for i in columns:
    dataSet[i] = pd.to_numeric(dataSet[i])
```

```python
print("Normal Red blood cells are : "+str(len(dataSet.rbc[(dataSet.rbc == 'normal') & (dataSet['class'] == 'ckd')])))
```

```python
print("Abnormal Red blood cells are : "+str(len(dataSet.rbc[(dataSet.rbc == 'abnormal') & (dataSet['class'] == 'ckd')])))
```

```python
normalCount = dataSet.rbc[(dataSet['class'] == 'ckd')].value_counts()
normalCount.plot(kind='bar', color = 'red')
plt.show()
```
