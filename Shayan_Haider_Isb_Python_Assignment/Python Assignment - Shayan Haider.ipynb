# 4. The dataset is not in clean form. Write a function that takes pandas dataframe as an input and replace these characters i-e ( "\t", " ", "?") with "nan" the function should apply this transformation to all columns in dataframe and for cross check display unique values of Diabetes Mellitus(dm).

```python
def data_set_clean(data_set):
    for i in data_set:
        data_set[i] = data_set[i].replace("" ,np.nan)
        data_set[i] = data_set[i].replace("?" , np.nan)
        data_set[i] = data_set[i].replace("nan" , np.nan)
        data_set[i] = data_set[i].replace("\t" , np.nan)    
        data_set[i] = data_set[i].replace("\tno" , "no")    
        data_set[i] = data_set[i].replace("\tyes" , "yes")
        data_set[i] = data_set[i].replace(" no" , "no")
        data_set[i] = data_set[i].replace(" yes" , "yes")


data_set_clean(data_set)

print data_set.dm.unique()
```

# 5. Change datatype to Numerical of following columns in Dataframe. columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

```python
def is_num(data_set):
    num_headers = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']
    for i in num_headers:
#         print i
        data_set[i] = pd.to_numeric(data_set[i],errors = "coerce")
        
is_num(data_set)
print data_set.age.dtypes       
```
