# The dataset is not in clean form. Write a function that takes pandas dataframe as an input and replace these characters i-e ( "\t", " ", "?") with "nan" the function should apply this transformation to all columns in dataframe and for cross check display unique values of Diabetes Mellitus(dm).

```python
def cleanDataset(data):
    return data.replace(r'[\t\s]', '', regex=True).replace("?",np.nan)

data = cleanDataset(data)
print "\n\nAfter cleaning: ", data.dm.unique()
```

# Change datatype to Numerical of following columns in Dataframe. columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']Â¶


```python
def convertToNumeric(column):
    for col in column:
        data[col] = pd.to_numeric(data[col], errors="coerce")
    return data
        
columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']
data = convertToNumeric(columns)
print "\n\nAter conversion to numeric"
for col in columns:
    print "%s is of %s type now " % (col, data[col].dtypes)
```

# Count the number of normal or abnormal red blood cells people having Chronic Kidney Disease(CKD).


```python
unique, counts = np.unique(data.loc[data['class'] == 'ckd', 'rbc'], return_counts=True)
D = dict(zip(unique, counts))
print "\n\nNormal: ", D['normal'] 
print "Abnormal: ", D['abnormal']
plt.bar(range(len(D) -1), [D['normal'], D['abnormal']], align='center')
plt.xticks(range(len(D) -1), ['normal', 'abnormal'])

plt.show()
```
